```python
# function to print a dictionary or a nested dictionary
def print_dict(d, indent=0):
    i = 1
    for key, value in d.items():
        print(' ' * indent + f"{i}) {key} :\t", end="")
        if isinstance(value, dict):
            print()
            print_dict(value, indent + 4)
        else:
            print(f"{value}\n")
        i += 1
```

```python
# writing a function that return the Fixed part of the file Description

def extract_fixed_file_description(data):
    """
    Extracts the fixed part (first 180 bytes) of the File Descriptor Record from an L3 file.
    Returns the extracted data as a dictionary.
    """ 
    file_descriptor = {
        "Record Number": struct.unpack(">I", data[0:4])[0],
        "Record Type and Subtype": struct.unpack(">I", data[4:8])[0],
        "Record Length": struct.unpack(">I", data[8:12])[0],
        "ASCII/EBCDIC Flag": data[12:14].decode(errors='ignore').strip(),
        "Spares": data[14:16],
        "Superstructure Format control Document Number": data[16:28].decode(errors='ignore').strip(),
        "Superstructure Control Document Revision Number": struct.unpack(">H", data[28:30])[0],
        "Superstructure Control Document Revision Letter": struct.unpack(">H", data[30:32])[0],
        "Software Release Number": data[32:44].decode(errors='ignore').strip(),
        "Sequence Number of File in Logical Volume": struct.unpack(">I", data[44:48])[0],
        "File Name": data[48:64].decode(errors='ignore').strip(),
        "Record Sequence and Location Type Flag": data[64:68].decode(errors='ignore').strip(),
        "Location of the start of the Sequence Number Field": struct.unpack(">Q", data[68:76])[0],
        "Length of Sequence Number Field": struct.unpack(">I", data[76:80])[0],
        "Record Code and Location Type Flag": data[80:84].decode(errors='ignore').strip(),
        "Location of Record Type Code Field": struct.unpack(">Q", data[84:92])[0],
        "Length of Record Type Code Field": struct.unpack(">I", data[92:96])[0],
        "Record Length and Location Type Flag": data[96:100].decode(errors='ignore').strip(),
        "Location of Record Length Field": struct.unpack(">Q", data[100:108])[0],
        "Length of Record Length Field": struct.unpack(">I", data[108:112])[0],
        "Flag for Data Interpretation in Descriptor": data[112:113].decode(errors='ignore').strip(),
        "Flag for Data Interpretation in Other Records": data[113:114].decode(errors='ignore').strip(),
        "Flag for Data Display in Descriptor": data[114:115].decode(errors='ignore').strip(),
        "Flag for Data Display in Other Records": data[115:116].decode(errors='ignore').strip(),
        "Spares": data[116:180]
    }
    return file_descriptor
```

```python
def extract_leader_file_descriptor(file_path):
    """
    Extracts the full Leader File Descriptor Record (6120 bytes)
    including the fixed file descriptor part and additional fields.
    """
    with open(file_path, "rb") as file:
        data = file.read(6120)  # Read the full Leader File Descriptor Record
    
    
    # converting binary to 6 byte numeric field
    def extract_numeric_field(start, end):
        """Extracts a numeric field of length 6 bytes as an integer."""
        return int.from_bytes(data[start:end], byteorder="big")

    leader_descriptor = {
        "Fixed part of the file descriptor": extract_fixed_file_description(data[:180]),            # Extract fixed part (1-180)
        "Number of Header Records": extract_numeric_field(180, 186),
        "Header Record Length": extract_numeric_field(186, 192),
        "Number of Ephemeris/Attitude Records": extract_numeric_field(192, 198),
        "Length of Ephemeris/Attitude Records": extract_numeric_field(198, 204),
        "Number of Calibration Records": extract_numeric_field(204, 210),
        "Length of Calibration Records": extract_numeric_field(210, 216),
        "Number of Histogram Records": extract_numeric_field(216, 222),
        "Length of Histogram Records": extract_numeric_field(222, 228),
        "Number of Map Projection Records": extract_numeric_field(228, 234),
        "Length of Map Projection Records": extract_numeric_field(234, 240),
        "Number of GCP Records": extract_numeric_field(240, 246),
        "Length of GCP Records": extract_numeric_field(246, 252),
        "Number of annotation Records": extract_numeric_field(252, 258),
        "Length of annotation Records": extract_numeric_field(248, 264),
        "Number of line and pixel LUT Records": extract_numeric_field(264, 270),
        "Length of line and pixel LUT Records": extract_numeric_field(270, 276),
        "Number of Attitude rate Records": extract_numeric_field(276, 282),
        "Length of Attitude rate Records": extract_numeric_field(282, 288),
        "Number of boundary Records": extract_numeric_field(288, 294),
        "Length of boundary Records": extract_numeric_field(294, 300),
        "Number of boundary annotations Records": extract_numeric_field(300, 306),
        "Length of boundary annotations Records": extract_numeric_field(306, 312),
        "Scene Identification Field Locator": data[324:340].decode(errors='ignore').strip(),
        "Mission Identification Field Locator": data[340:356].decode(errors='ignore').strip(),
        "Sensor Identification Field Locator": data[356:372].decode(errors='ignore').strip(),
        "Exposure Date & Time Locator": data[372:388].decode(errors='ignore').strip(),
        "Geographic Field Reference Locator": data[388:404].decode(errors='ignore').strip(),
        "Image Processing Performed Field Locator": data[404:420].decode(errors='ignore').strip(),
        "Imagery Format Indicator Locator": data[420:436].decode(errors='ignore').strip(),
        "Band Indicator Locator": data[436:452].decode(errors='ignore').strip(),
        # "Spares": data[452:6120]                          # as very large amount of unneseccary data
    }
    return leader_descriptor

# Testing usage of function using print_dict function created above
leader_descriptor_data = extract_leader_file_descriptor("LEADER.L-3")

print_dict(leader_descriptor_data)
```

```python
# working on with attitude records
def extract_ephemeris_attitude_record(file_path):
    """
    Extracts the Ephemeris/Attitude Record (6120 bytes) (Table 2.4.3.3) from the Leader File.
    Correctly handles floating-point fields and repetitions.
    """
    with open(file_path, "rb") as file:
        file.seek(12240)  # Move to the start of the Ephemeris/Attitude Record (after Header Record)
        data = file.read(6120)  # Read 6120 bytes (Ephemeris/Attitude Record)
    
    ephemeris_record = {
        "Record Number": struct.unpack(">I", data[0:4])[0],
        "Record Type and Subtype": struct.unpack(">I", data[4:8])[0],
        "Record Length": struct.unpack(">I", data[8:12])[0],
        "Record Sequence Number": struct.unpack(">I", data[12:16])[0],
        "Number of Ephemeris/Attitude Values": struct.unpack(">I", data[16:20])[0],
        
        # Satellite Position XYZ (First Set) with F12.4 (12 bytes)
        "Satellite Position X (km, precision)": (
            struct.unpack(">d", data[20:28])[0],  # First 8 bytes
            struct.unpack(">f", data[28:32])[0] if len(data[28:32]) == 4 else None  # Last 4 bytes with safety check
        ),
        "Satellite Position Y (km, precision)": (
            struct.unpack(">d", data[500:508])[0],
            struct.unpack(">f", data[508:512])[0] if len(data[508:512]) == 4 else None
        ),
        "Satellite Position Z (km, precision)": (
            struct.unpack(">d", data[980:988])[0],
            struct.unpack(">f", data[988:992])[0] if len(data[988:992]) == 4 else None
        ),
        
        # Velocity Vectors XYZ (First Set) with F12.7 (12 bytes)
        "Velocity Vector X (km/sec, precision)": (
            struct.unpack(">d", data[1460:1468])[0],
            struct.unpack(">f", data[1468:1472])[0] if len(data[1468:1472]) == 4 else None
        ),
        "Velocity Vector Y (km/sec, precision)": (
            struct.unpack(">d", data[1940:1948])[0],
            struct.unpack(">f", data[1948:1952])[0] if len(data[1948:1952]) == 4 else None
        ),
        "Velocity Vector Z (km/sec, precision)": (
            struct.unpack(">d", data[2420:2428])[0],
            struct.unpack(">f", data[2428:2432])[0] if len(data[2428:2432]) == 4 else None
        ),
        
        # Satellite Altitude with F16.7 (16 bytes)
        "Satellite Altitude (m, degrees)": struct.unpack(">2d", data[4012:4028]),
        
        # Scene Centre Time
        "Scene Centre Time": data[4028:4054].decode(errors='ignore').strip(),
        "Ephemeris/Attitude Start Time": data[4054:4066].decode(errors='ignore').strip(),
        "Time Interval for Ephemeris/Attitude (s)": struct.unpack(">f", data[4066:4072])[0] if len(data[4066:4072]) == 4 else None,
        
        # Yaw, Roll, Pitch with F10.5 (10 bytes, handling as float)
        "Yaw (deg.)": struct.unpack(">f", data[4100:4110])[0] if len(data[4100:4110]) == 4 else None,
        "Roll (deg.)": struct.unpack(">f", data[4500:4510])[0] if len(data[4500:4510]) == 4 else None,
        "Pitch (deg.)": struct.unpack(">f", data[4900:4910])[0] if len(data[4900:4910]) == 4 else None,
        
        # Scene Centre Attitude Angles
        "Scene Centre Yaw (deg.)": struct.unpack(">f", data[5300:5310])[0] if len(data[5300:5310]) == 4 else None,
        "Scene Centre Roll (deg.)": struct.unpack(">f", data[5310:5320])[0] if len(data[5310:5320]) == 4 else None,
        "Scene Centre Pitch (deg.)": struct.unpack(">f", data[5320:5330])[0] if len(data[5320:5330]) == 4 else None,
        
        # Swath Flag
        "Swath Flag": data[6118:6120].decode(errors='ignore').strip(),
    }
    # finally return 
    return ephemeris_record
```

## **Common Feilds in Both the text file**

```python
# creating a function that will extract all the fields from the text files 
def extract_fields(file_path):
    """
    Extracts field names from a structured text file.

    :param file_path: Path to the text file
    :return: Set of field names
    """
    fields = set()
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            parts = line.strip().split(":")             # Split at colon
            if len(parts) > 1:                          # Ensure it's a key-value pair
                field_name = parts[0].strip()
                fields.add(field_name)
    return fields
# function to compare the two files.
def compare_files(file1, file2):
    """
    Compares two text files and finds common fields.

    :param file1: Path to first file
    :param file2: Path to second file
    :return: Set of common fields
    """
    fields1 = extract_fields(file1)
    fields2 = extract_fields(file2)

    common_fields = fields1.intersection(fields2)

    return common_fields

# Testing usage of function created above
file1 = "leader_extracted.txt"
file2 = "imagery_extracted.txt"
common_fields = compare_files(file1, file2)
print("Common Fields Found: (ignore numbering)\n")
for field in sorted(common_fields):
    print(field)
```
